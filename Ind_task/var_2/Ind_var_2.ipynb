## Досліджуємо вплив використання соціальних мереж на емоційний стан користувачів  
На основі даних про активність користувачів у соціальних мережах та їх домінантний емоційний стан вам необхідно дослідити можливість прогнозування емоційного стану користувачів залежно від характеру їхньої активності.  

**Опис датасету:**  
User_ID: Унікальний ідентифікатор користувача.  
Age: Вік користувача.  
Gender: Стать користувача (Female, Male, Non-binary).  
Platform: Соціальна мережа, якою користується користувач (Instagram, Twitter, Facebook, LinkedIn, Snapchat, Whatsapp, Telegram).  
Daily_Usage_Time (minutes): Щоденний час користування платформою в хвилинах.  
Posts_Per_Day: Кількість постів за день.  
Likes_Received_Per_Day: Кількість отриманих вподобайок за день.  
Comments_Received_Per_Day: Кількість отриманих коментарів за день.  
Messages_Sent_Per_Day: Кількість надісланих повідомлень за день.  
Dominant_Emotion: Домінантний емоційний стан користувача протягом дня (наприклад, Happiness, Sadness, Anger, Anxiety, Boredom, Neutral).  

**Завдання**
На основі аналізу та обробки даних у датасеті, визначте залежність між активністю користувачів у соцмережах та їхнім емоційним станом( <span style="color:red"> Особливу увагу зверніть на колонку Dominant_Emotion, яка містить декілька можливих значень емоційного стану користувача. Також зверніть увагу, що дану змінну варто перекодувати за допомогою label encoding або в ідеалі ordinal encoding</span>):  
1. Виконайте попередню обробку даних, враховуючи можливі проблеми з даними.
2. Проведіть дослідження даних та виявлення закономірностей у поведінці користувачів. За результатами проведеного дослідження сформулюйте висновки.  
- Яка соціальна мережа є найпопулярнішою серед користувачів загалом та в розрізі віку/статі?  
- Чи спостерігається зв’язок між щоденним часом використання соцмереж і частотою постів/повідомлень/вподобайок?  
- Чи мають користувачі з домінантним станом "Anxiety", "Anger" або "Sadness" відмінності в активності порівняно з "Happiness" чи "Neutral"?  
- Чи відрізняється характер активності на різних платформах (наприклад, оберіть Telegram vs Instagram)?  
- Чи існують значущі відмінності у середніх показниках активності між групами за емоційним станом?  
- Побудуйте матрицю кореляцій між кількісними змінними: які дії в соцмережах найбільше взаємопов’язані?  
3. Оберіть алгоритм машинного навчання для прогнозування домінантного емоційного стану користувача, враховуючи особливості цільової змінної.
4. Навчіть кілька алгоритмів для прогнозування домінантного емоційного стану користувача та порівняйте їх якість. Оберіть найкращий алгоритм для подальшої роботи.  
5. Використайте обраний алгоритм для прогнозування емоційного стану для вибраних користувачів (інформація про них міститься в 'users_to_predict.csv').
