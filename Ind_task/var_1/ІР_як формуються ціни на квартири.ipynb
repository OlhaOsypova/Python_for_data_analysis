## Як формуються ціни на квартири?  
Датасет **prices.csv** містить пропозиції квартир з 15 найбільших міст Польщі (Варшава, Лодзь, Краків, Вроцлав, Познань, Гданськ, Щецин, Бидгощ, Люблін, Катовіце, Білосток, Ченстохова). Дані взяті з місцевих сайтів з продажу квартир та охоплюють період за жовтень 2024 року:  
- city - назва міста, де знаходиться квартира    
- type - тип будівлі  
- squareMeters - площа квартири в квадратних метрах  
- rooms - кількість кімнат у квартирі  
- floor / floorCount - поверх, на якому розташована квартира, і загальна кількість поверхів у будинку  
- buildYear - рік побудови будівлі  
- centreDistance - відстань від центру міста в км  
- poiCount - кількість об'єктів в радіусі 500 м від квартири (школи, поліклініки, пошти, дитячі садки, ресторани, коледжі, аптеки)   
- schoolDistance', 'clinicDistance', 'postOfficeDistance', 'kindergartenDistance', 'restaurantDistance', 'collegeDistance', 'pharmacyDistance' - відстань до найближчої точки інтересу (школи, поліклініки, пошти, дитячі садки, ресторани, коледжі, аптеки)  
- ownership - вид володіння майном  
- condition - стан квартири   
- hasParkingSpace', 'hasBalcony', 'hasElevator', 'hasSecurity', 'hasStorageRoom' - чи має нерухомість ключові характеристики, такі як виділене місце для паркування, балкон, ліфт, охорона, комора  
- price - ціна квартири в польських злотих

- Аналіз розподілу кількості кімнат та площі у квартирах:  
Дослідіть розподіл кількості кімнат у квартирах. Порівняйте середні ціни на квартири з різною кількістю кімнат.  
Дослідіть розподіл квартир за площею.  
Проаналізуйте, чи є залежність між кількістю кімнат, площею та ціною квартири.  
- Аналіз залежності ціни від відстані до центру міста:  
Порівняйте розподіл цін на квартири в залежності від відстані до центру міста.  
Дослідіть зв'язок між відстанню до центру міста та ціною квартири в різних містах.  
- Аналіз впливу різних зручностей на ціну квартир:  
Дослідіть, як наявність різних зручностей (наприклад, балкон, ліфт, виділене місце для паркування) впливає на ціну квартир.  
Порівняйте розподіл цін на квартири з різними зручностями та без них.  
- Аналіз розподілу років побудови будинків:  
Вивчіть розподіл років побудови будинків, у яких розташовані квартири.  
Дослідіть, чи є зв'язок між роком побудови будинку та ціною квартири.
